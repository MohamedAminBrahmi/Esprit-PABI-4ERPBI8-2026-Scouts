```python
import pandas as pd
from deep_translator import GoogleTranslator, exceptions

INPUT_FILE = r"C:\Users\MSI\OneDrive\Desktop\ESPRIT\Youth Leadership & Outdoor Education\Scouts\Détails des Coûts de Camp (Dinars).xlsx"
OUTPUT_FILE = "Détails des Coûts de Camp (Dinars)_translated.xlsx"
COLUMNS_TO_TRANSLATE = None  # e.g., ["Description", "Notes"]

translator = GoogleTranslator(source='fr', target='en')

def safe_translate(text):
    if not isinstance(text, str) or not text.strip():
        return text
    try:
        return translator.translate(text)
    except (exceptions.NotValidPayload,
            exceptions.NotValidLength,
            exceptions.TranslationNotFound,
            Exception):
        return text  # keep original on any error

# Load data
df = pd.read_excel(INPUT_FILE)

# Translate headers
translated_columns = []
for col in df.columns:
    translated_columns.append(safe_translate(col))
df.columns = translated_columns

# Columns to translate (data)
cols = COLUMNS_TO_TRANSLATE or [c for c in df.columns if df[c].dtype == object]

for col in cols:
    df[col] = df[col].apply(safe_translate)

# Save
df.to_excel(OUTPUT_FILE, index=False)
print(f"Saved translated file to: {OUTPUT_FILE}")
```
